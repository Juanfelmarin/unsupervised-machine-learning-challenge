```python
#import dependencies
import pandas as pd
from pathlib import Path
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
```

```python
#read csv
unsupervised_df = pd.read_csv("/users/juanmarin/desktop/Data_bootcamp/unsupervised-machine-learning-challenge/Starter_Code_20/myopia.csv")
unsupervised_df
```

```python
#dropping column
myopia_df = unsupervised_df.drop('MYOPIC', axis=1)
myopia_df
```

```python
#standerdize
scaled_myopia_df = StandardScaler().fit_transform(myopia_df)
```

```python
#initialize PCA and setting to 90% of explained variance
pca = PCA(n_components=0.9)
myopia_pca = pca.fit_transform(scaled_myopia_df)
```

```python
# Create a new dataframe for t-SNE
myopia_pca_df = pd.DataFrame(
    data=myopia_pca)
myopia_pca_df
```

```python
#Further reduce the dataset dimensions with t-SNE.
tsne = TSNE(learning_rate=200)
myopia_tsne_df = tsne.fit_transform(myopia_pca_df)
```

```python
#define x and y for plotting
myopia_pca_df['x'] = myopia_tsne_df[:,0]
myopia_pca_df['y'] = myopia_tsne_df[:,1]
```

```python
#plot the scatterplot
plt.scatter(myopia_pca_df['x'], myopia_pca_df['y'])
```

```python
#k means 
inertia = []
k = list(range(1, 10))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(myopia_pca_df)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()

# Plot the elbow curve to find the best candidate(s) for k
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,10))
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve')
plt.show()
```

# According to this graph it seems like the elbow curve is at 5.

```python
#straight from past kmeans solved assignment
#start kmeans
def get_clusters(k, data):
    # Initialize the K-Means model
    model = KMeans(n_clusters=k, random_state=0)

    # Train the model
    model.fit(data)

    # Predict clusters
    predictions = model.predict(data)

    # Create return DataFrame with predicted clusters
    data["class"] = model.labels_

    return data
```

```python
#get clusters and used 5 becuase it was the elbow curve
clusters = get_clusters (5, myopia_pca_df )
clusters
```

```python
#show my clusters in scatterplot
def show_clusters(myopia_pca_df):
    plt.scatter(myopia_pca_df[0], myopia_pca_df[1], c=myopia_pca_df["class"])
    plt.xlabel('0')
    plt.ylabel('1')
    plt.show()
```

```python
show_clusters(clusters)
```

# According to these results it is possible to cluster the patients in five clusters, since there are five distinct clusters in the scatterplot.
